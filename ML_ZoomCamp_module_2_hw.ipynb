#Creating validation framework

```python
n = len(req_df)
n_val = int(n*0.2)
n_test = int(n*0.2)
n_train = n-(n_val+n_test)

# n_val, n_test, n_train

idx = np.arange(n)
np.random.seed(42)
np.random.shuffle(idx)

# df_train = req_df.iloc[:n_train]
# df_val = req_df.iloc[n_train:n_val+n_train]
# df_test = req_df.iloc[n_train+n_val:]

df_train = req_df.iloc[idx[:n_train]]
df_val = req_df.iloc[idx[n_train:n_val+n_train]]
df_test = req_df.iloc[idx[n_train+n_val:]]
# df_train.head()

df_train = df_train.reset_index(drop = True)
df_val = df_val.reset_index(drop = True)
df_test = df_test.reset_index(drop = True)
df_train.head()
```

```python
y_train = np.log1p(df_train.fuel_efficiency_mpg.values)
y_val = np.log1p(df_val.fuel_efficiency_mpg.values)
y_test = np.log1p(df_test.fuel_efficiency_mpg.values)

del df_train['fuel_efficiency_mpg']
del df_val['fuel_efficiency_mpg']
del df_test['fuel_efficiency_mpg']
```

```python
def train_linear_regression(X,y):
  ones = np.ones(X.shape[0])
  X = np.column_stack([ones, X])

  XTX = X.T.dot(X)
  XTX_inv = np.linalg.inv(XTX)
  w = XTX_inv.dot(X.T).dot(y)

  return w[0], w[1:]

def prepare_X(df, fill_na):
    df = df.fillna(fill_na)
    X = df.values
    return X

def rmse(y_val,y_pred):
  error = y_val - y_pred
  s_error = (error**2).mean()
  return np.sqrt(s_error)

```

With null values as 0

```python
X_train = prepare_X(df_train, 0)
w_0 , w = train_linear_regression(X_train, y_train)

X_val = prepare_X(df_val, 0)
y_pred = w_0 + X_val.dot(w)

round(rmse(y_val,y_pred),2)
```

with mean is null values


```python
mean_null = df_train.horsepower.mean()
X_train = prepare_X(df_train, mean_null)
w_0 , w = train_linear_regression(X_train, y_train)

X_val = prepare_X(df_val, mean_null)
y_pred = w_0 + X_val.dot(w)

round(rmse(y_val,y_pred),2)
```
